```python
import requests
import bs4
import pandas as pd
import datetime 
```

```python
import pymysql 
from sqlalchemy import create_engine 
```

```python
# Selenium 셀레늄으로 Chrome 웹 브라우저 띄우기 

from selenium import webdriver

from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager

chromeOption = Options()

chromeOption.add_argument("window-size=1920X1080")

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install() ),
                          options=chromeOption)
```

```python
# 웹페이지 이동하기
targetUrl = "https://www.naver.com/"
driver.get(targetUrl)
```

```python
# 버튼 클릭하기 

BtnXpath = '//*[@id="NM_FAVORITE"]/div[1]/ul[2]/li[2]/a'
EconomyBtn = '/html/body/section/header/div[2]/div/div/div[1]/div/div/ul/li[3]/a/span'
from selenium.webdriver.common.by import By

driver.find_element(By.XPATH, BtnXpath).click()
driver.implicitly_wait(3)
```

```python
driver.find_element(By.XPATH, EconomyBtn).click()
```

```python
# 수집한 날짜 정보 저장
currentDate=datetime.datetime.now()

timeKey = currentDate.strftime("%Y_%m_%d")
```

```python
# 페이지 소스받아와서 html로 변경해주기
textObj = driver.page_source

htmlObj=bs4.BeautifulSoup(textObj, "html.parser")


# 네이버 경제면 헤드라인 뽑아오기 
mainContent = htmlObj.find(id = "main_content")


mainContentInfo =mainContent.findAll(name = "li", attrs = {"class" : "cluster_item"})
mainContentInfo[0].find(attrs = {"class" :"cluster_text"}).find (name ="a")

headlineInfoList = []

for i in range (0, len(mainContentInfo)) :
    try : 
        headline = mainContentInfo[i].find(attrs = {"class" :"cluster_text"}).find (name ="a")
        headlineInfoList.append(headline)
    except Exception as e : 
        print(e)

headLineList =[]

for i in range(0, len(headlineInfoList)) :    
    mainHead = headlineInfoList[i].text                
    url = headlineInfoList[i].attrs['href'] 
    
    headLineList.append([mainHead,url] )

newHeadlineData = pd.DataFrame (headLineList)
newHeadlineData.columns = ["헤드라인", "URL 링크"]
newHeadlineData
```

```python
newHeadlineData["TIME_KEY"] = timeKey # 수집시간 정보 데이터에 추가하기

columnArrange=["TIME_KEY", "URL 링크", "헤드라인"]

newHeadlineData = newHeadlineData.loc[:,columnArrange ] 
```

```python
# DB 커넥션 열기
inId = "root"
inPw = "0000"
inIp = "127.0.0.1"
inPort ="3306"
inDbName = "web"
engine_mariadb = create_engine('mysql+pymysql://{}:{}@{}:{}/{}'.format(inId,inPw,inIp,inPort,inDbName))
resultname = "KOPO_HEADLINENEWS"
newHeadlineData.to_sql(name=resultname, con = engine_mariadb , if_exists = "append" , index = False) # if_exists = append로 있는 정보 계속 더하기  
```

```python
pd.read_sql_query("select * from KOPO_HEADLINENEWS", con=engine_mariadb) # mariaDB sql로 데이터 저장
```

```python
newHeadlineData.to_csv("./news.csv", encoding="ms949") #엑셀파일로 저장 - 자동화 
```
